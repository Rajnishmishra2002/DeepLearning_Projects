```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from mlxtend.plotting import plot_decision_regions
from matplotlib.colors import ListedColormap
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles

import seaborn as sns
```

```python
X,y =make_circles(n_samples=100,noise=0.1,random_state =1)
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=y)
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size =0.2,random_state =3,)
```

```python
model =Sequential()
model.add(Dense(256,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
```

```python
history =model.fit(X_train,y_train,validation_data =(X_test,y_test)  ,epochs=3500,verbose =1)
```

```python
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label ='test')
plt.legend()
plt.show()
```

```python
plot_decision_regions(X_test,y_test.ravel(),clf=model,legend=2)
```

Early Stopping

```python
model =Sequential()
model.add(Dense(256,input_dim=2,activation ='relu'))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
```

```python
callback =EarlyStopping(monitor="val_loss",min_delta=0.000001,
                        patience=20,verbose =1,mode="auto",baseline=None,restore_best_weights=False)
```

```python
history =model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=3000,callbacks=callback)
```

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```
